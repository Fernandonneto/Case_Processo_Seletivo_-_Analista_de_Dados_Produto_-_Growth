### A) Distribuição por faixa etária e renda:

```python
# Distribuição por faixa etária e renda: Gráfico de barras (value_counts) para as variáveis (age_range e salary_range)
```

```python
# Gráfico - Faixa Etária

contagem = df['age_range'].value_counts()

# Define as cores
cores = [PALETA["cinza"]] * len(contagem)
for i in range(min(2, len(contagem))):
    cores[i] = PALETA["azul"]

plt.figure(figsize=(8, 5))

ax = contagem.plot(
    kind='bar',
    color=cores,
    edgecolor='white'
)

plt.title('Distribuição dos Usuários por Faixa Etária', fontsize=14)
plt.xlabel('')
plt.ylabel('Número de Usuários')

ax.grid(axis='y', linestyle='--', alpha=0.30)
ax.grid(axis='x', visible=False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Gráfico - Faixa de Renda
contagem = df['salary_range'].value_counts()

cores = [PALETA["cinza"]] * len(contagem)
for i in range(min(2, len(contagem))):
    cores[i] = PALETA["azul"]

plt.figure(figsize=(8, 5))

ax = contagem.plot(
    kind='bar',
    color=cores,
    edgecolor='white'
)

plt.title('Distribuição dos Usuários por Faixa de Renda', fontsize=14)
plt.xlabel('')
plt.ylabel('Número de Usuários')

ax.grid(axis='y', linestyle='--', alpha=0.30)
ax.grid(axis='x', visible=False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Insights:**

**Distribuição de usuários por faixa etária:** Observa-se que a categoria "Não informado" representa a maior parte dos registros da base. Durante a etapa de tratamento dos dados, verificou-se que a ausência dessa informação ocorre simultaneamente com outras variáveis demográficas, indicando uma forte hipótese de que esses registros correspondam a usuários que realizaram partidas, mas não possuem um perfil cadastrado no sistema. Essa hipótese será explorada mais detalhadamente nas análises de retenção e comportamento dos usuários. Entre os usuários que possuem a faixa etária informada, destaca-se a predominância da faixa de 25 a 34 anos, seguida por 35 a 44 anos, enquanto as faixas 18 a 24 anos e 45+ apresentam menor participação.

**Distribuição de usuários por faixa de renda:** A variável salary_range apresenta comportamento semelhante ao observado em age_range, com predominância da categoria "Não informado". Esse padrão reforça a hipótese de que uma parcela significativa dos usuários ativos na plataforma não possui informações demográficas cadastradas. Entre os usuários que informaram a faixa salarial, observa-se maior concentração nas rendas entre R$ 2 mil e R$ 6 mil, enquanto as demais faixas apresentam menor representatividade. Dessa forma, as análises relacionadas ao perfil socioeconômico devem considerar que uma parcela expressiva da base não disponibilizou essas informações.

### B) Ecossistema tecnológico:

```python
# Ecossistema tecnológico: Tabela de contingência (pd.crosstab) entre o dispositivo de sessão (device) e o 
# dispositivo principal (primary_device)
```

```python
# Frequências absolutas
tabela_freq = pd.crosstab(
    index=df['device'],
    columns=df['primary_device'],
    margins=True,
    margins_name='Total'
)

tabela_freq.index.name = 'Dispositivo da Sessão'
tabela_freq.columns.name = 'Dispositivo Principal'

# Percentuais por linha
tabela_pct = (
    pd.crosstab(
        index=df['device'],
        columns=df['primary_device'],
        normalize='index'
    ) * 100
).round(1)

tabela_pct.index.name = 'Dispositivo da Sessão'
tabela_pct.columns.name = 'Dispositivo Principal'

# Exibição
print("Tabela 1 - Frequência absoluta")
display(tabela_freq)

print("\nTabela 2 - Percentual por dispositivo da sessão (%)")
display(tabela_pct)
```

**Insights:**

A tabela de contingência evidencia uma elevada consistência entre o dispositivo principal informado pelo usuário e o dispositivo utilizado durante a sessão. Entre os usuários cujo dispositivo principal é Android, aproximadamente 65,7% das sessões também foram realizadas em dispositivos Android, enquanto não foram observadas sessões em iOS. Da mesma forma, 68,3% dos usuários cujo dispositivo principal é iOS jogaram utilizando esse mesmo sistema operacional, sem registros de utilização em dispositivos Android.

Além disso, observa-se uma parcela expressiva de usuários classificados como "Não informado" na variável de dispositivo principal (34,3% das sessões em Android e 31,7% das sessões em iOS). Esse comportamento reforça a hipótese levantada durante o tratamento dos dados de que parte dos jogadores utiliza a plataforma sem possuir um perfil cadastrado ou com informações demográficas incompletas.

Outro ponto relevante é a ausência de migração entre os ecossistemas Android e iOS, sugerindo que os usuários tendem a permanecer utilizando o mesmo sistema operacional para acessar o jogo. Esse comportamento pode ser considerado em futuras estratégias de comunicação, desenvolvimento de funcionalidades e campanhas específicas para cada plataforma.

### C) Perfil profissional:

```python
# Perfil profissional: Contagem de frequência das variáveis (job_role) e (sector) para entender quais áreas 
# de trabalho mais jogam Palavritas
```

```python
# Frequência das variáveis - Cargo/Função
freq_job = (
    df['job_role']
    .value_counts(dropna=False)
    .rename_axis('Cargo/Função')
    .reset_index(name='Frequência')
)

freq_job['Percentual (%)'] = (
    freq_job['Frequência'] / freq_job['Frequência'].sum() * 100
).round(1)

print("Tabela 1 - Distribuição dos usuários por cargo/função")
display(freq_job)

# Frequência das variáveis - Setor de atuação
freq_sector = (
    df['sector']
    .value_counts(dropna=False)
    .rename_axis('Setor')
    .reset_index(name='Frequência')
)

freq_sector['Percentual (%)'] = (
    freq_sector['Frequência'] / freq_sector['Frequência'].sum() * 100
).round(1)

print("\nTabela 2 - Distribuição dos usuários por setor de atuação")
display(freq_sector)
```

**Insights:**

Os perfis profissionais dos usuários são bastante diversificados. Em ambas as variáveis job_role e sector, a categoria "Não informado" representa 32,8% dos registros, reforçando a hipótese de que uma parcela significativa joga sem possuir cadastro.

Entre os que preencheram o cargo, destacam-se Diretores (7,6%), Analistas (6,9%), Coordenadores (6,4%), Gerentes (6,0%) e Estudantes (5,8%), mostrando uma forte presença de profissionais de gestão e estudantes. Nota-se também um público especializado na área de dados, com cargos como Analista, Engenheiro e Desenvolvedor de Dados.

No setor de atuação, Tecnologia lidera com 14,8%, seguida por Outros (12,6%), Finanças (9,6%) e Educação (8,5%). Isso comprova que, embora o setor tecnológico seja predominante, a plataforma atrai usuários de diversos segmentos do mercado (como Saúde, Varejo, Marketing e Direito).
