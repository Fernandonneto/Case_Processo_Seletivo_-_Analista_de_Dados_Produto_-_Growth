## 1º - Quem são os usuários?

Mapeamento dos perfis demográficos e técnologicos.

### A) Distribuição por faixa etária e renda:

Essa hipótese será explorada mais detalhadamente nas análises de retenção e comportamento dos usuários.
```python
# Frequências absolutas
tabela_freq = pd.crosstab(
    index=df['device'],
    columns=df['primary_device'],
    margins=True,
    margins_name='Total'
)

tabela_freq.index.name = 'Dispositivo da Sessão'
tabela_freq.columns.name = 'Dispositivo Principal'

# Percentuais por linha
tabela_pct = (
    pd.crosstab(
        index=df['device'],
        columns=df['primary_device'],
        normalize='index'
    ) * 100
).round(1)

tabela_pct.index.name = 'Dispositivo da Sessão'
tabela_pct.columns.name = 'Dispositivo Principal'

# Exibição
print("Tabela 1 - Frequência absoluta")
display(tabela_freq)

print("\nTabela 2 - Percentual por dispositivo da sessão (%)")
display(tabela_pct)
```

### C) Análise das palavras:

```python
def plot_top_loss_words(df, top_n=10):
    """
    Gera automaticamente o ranking das palavras com maior taxa de derrota (Lose)
    e plota o gráfico já estilizado
    """

    # Padroniza resultado
    df = df.copy()
    df['result'] = df['result'].str.lower()

    # Calcula taxa de derrota
    loss_rate = (
        df.groupby('word')['result']
        .apply(lambda x: (x == 'lose').mean())
        .sort_values(ascending=False)
        .head(top_n)
    )

    # Define cores (paleta padrão)
    cores = [PALETA["cinza"]] * len(loss_rate)
    cores[0] = PALETA["azul"]
    cores[1] = PALETA["azul"]

    # Plot
    ax = loss_rate.sort_values().plot(
        kind='barh',
        color=cores[::-1],  # mantém ranking correto visualmente
        edgecolor='white',
        figsize=(10, 5)
    )

    ax.set_title('Top Palavras com Maior Taxa de Derrota', fontsize=14)
    ax.set_xlabel('Taxa de derrota')
    ax.set_ylabel('')

    ax.grid(axis='x', linestyle='--', alpha=0.6)
    ax.grid(axis='y', visible=False)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.tight_layout()
    plt.show()

plot_top_loss_words(df)
```

**Insights:**

A análise do ranking das 10 palavras com maior taxa de derrota revela um padrão claro: as 6 palavras mais desafiadoras para os usuários (PAZÃO, DANÇA, NIXÃO, HERÓI, ORAÇÃO e CIÚME) possuem acentos ou sinais gráficos, registrando uma taxa de derrota próxima a 80%. Essa observação é um insumo valioso de game design, pois comprova que a presença de caracteres especiais eleva drasticamente a dificuldade do jogo. 

Recomendação estratégica: Esse conhecimento pode ser utilizado estrategicamente para calibrar a dificuldade do quiz, alternando entre palavras mais simples ou mais complexas a depender do objetivo de engajamento da plataforma.

## 3º - Quem retorna?

Comparações entre quem voltou e quem não voltou.

### A) Engajamento vs. Performance:

```python
# Padroniza a variável
df['played_next_day'] = (
    df['played_next_day']
    .astype(str)
    .str.lower()
    .map({'true': 1, 'false': 0})
)

# Agrupa por streak
resumo = (
    df.groupby('streak_day')
      .agg(
          taxa_retorno=('played_next_day', 'mean'),
          n=('played_next_day', 'count')
      )
      .reset_index()
)

# Intervalo de confiança de 95% (aproximação normal)
resumo['erro'] = 1.96 * np.sqrt(
    (resumo['taxa_retorno'] * (1 - resumo['taxa_retorno'])) / resumo['n']
)

resumo['lim_inf'] = (resumo['taxa_retorno'] - resumo['erro']).clip(lower=0)
resumo['lim_sup'] = (resumo['taxa_retorno'] + resumo['erro']).clip(upper=1)

# Gráfico
plt.figure(figsize=(10,5))

plt.plot(
    resumo['streak_day'],
    resumo['taxa_retorno'],
    marker='o',
    linewidth=2,
    color=PALETA["azul"],
    label='Probabilidade de retorno'
)

plt.fill_between(
    resumo['streak_day'],
    resumo['lim_inf'],
    resumo['lim_sup'],
    color=PALETA["azul"],
    alpha=0.20,
    label='IC 95%'
)

plt.title('Probabilidade de Retorno no Dia Seguinte por Sequência', fontsize=14)
plt.xlabel('Dias consecutivos (Streak)')
plt.ylabel('Probabilidade de retorno')

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.grid(axis='x', visible=False)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.ylim(0, 1.05)
plt.legend(frameon=False)

plt.tight_layout()
plt.show()
```

**Insights:**

A análise da probabilidade de retorno por sequência (Streak) revela que o hábito de jogo se fortalece progressivamente até o 4º dia consecutivo, onde a chance de retorno atinge o pico estável de 30%.

Contudo, a partir do 5º dia, observa-se um aumento expressivo na volatilidade dos dados (alargamento do Intervalo de Confiança), sugerindo que manter o engajamento a longo prazo torna-se imprevisível e desafiador para o usuário.

Recomendação estratégica: Para mitigar a fadiga do hábito e evitar o desabamento da retenção observado após o 7º dia, a plataforma deve implementar gatilhos de incentivo (notificações com bônus, desafios especiais de meio de semana ou recompensas por manter o streak) focados especialmente em usuários que estão cruzando a marca do 4º e do 5º dia de jogo consecutivo.

**Insights:**

A análise de retenção em 30 dias revela que a fidelidade á plataforma é homogênea entre as gerações, estabilizando-se entre 31% para todas as faixas etárias. Esse comportamento uniforme comprova o apelo universal do quiz. O grupo "Não informado" (jogadores anônimos) acompanha exatamente o mesmo patamar de fidelidade, confirmando que os usuários sem cadastro possuem alto valor e são tão fiéis quanto os cadastrados.

Por outro lado, quando avaliamos o comportamento por turno, há variações importantes: os usuários matutinos (Morning) são os mais fiéis, registrando a maior taxa de retenção da base com cerca de 37%, o que sugere que o jogo funciona melhor quando integrado à rotina matinal. Em contrapartida, o período da noite (Night) apresenta o menor índice de retenção, ficando em aproximadamente 29%. Esse dado traz um contraponto crucial: embora a noite registre o maior volume absoluto de acessos, ela é a menos eficiente para fidelizar a longo prazo.

Recomendação estratégica: Como a taxa de inatividade geral atinge 69%, as ações de melhoria devem focar em mecânicas globais de retenção (como sistemas de níveis, conquistas ou um modo competitivo), em vez de campanhas segmentadas por idade. Além disso, para transformar os jogadores casuais da noite em usuários recorrentes, a plataforma pode disparar notificações push logo no início do dia (ex: entre 6h e 8h), incentivando a criação de um hábito matutino saudável na base.

### C) Matriz de correlação númerica:

```python
# Seleciona apenas as variáveis numéricas
variaveis = [
    'attempts',
    'time_to_complete_sec',
    'session_hour',
    'streak_day',
    'food_delivery_freq_week',
    'correct_letters',
    'correct_positions'
]

# Calcula a matriz de correlação
corr = df[variaveis].corr(method='pearson')

# Gráfico
plt.figure(figsize=(9, 7))

ax = sns.heatmap(
    corr,
    annot=True,              # Exibe os coeficientes
    fmt=".2f",               # Duas casas decimais
    cmap="Blues",            # Paleta em tons de azul
    vmin=-1,
    vmax=1,
    center=0,
    linewidths=0.5,
    square=True,
    cbar_kws={'label': 'Correlação'}
)

plt.title('Matriz de Correlação das Variáveis Numéricas', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

**Insights:**

A Matriz de Correlação das variáveis numéricas revela como a dinâmica do jogo influencia as métricas de desempenho dos usuários:

- Fortes Correlações Negativas: A relação mais intensa ocorre entre tentativas (attempts) e letras no lugar certo (correct_positions), registrando -0,65. Isso indica que quanto mais rápido o jogador descobre a posição correta das letras, de menos tentativas ele precisa para vencer. Da mesma forma, a correlação de -0,47 entre letras no lugar certo e letras certas em posição errada (correct_letters) mostra que preencher as posições corretas diminui a necessidade de ficar testando novas letras na sorte.

- Correlações Positivas Fracas: Nota-se uma relação de 0,26 entre tentativas e letras certas em posição errada, o que é perfeitamente lógico: jogadores que gastam mais tentativas naturalmente acabam digitando e descobrindo um volume maior de letras certas ao longo da partida. Além disso, a sutil correlação de 0,18 entre tentativas e hora da sessão (session_hour) indica uma tendência muito leve de as partidas jogadas mais tarde no dia exigirem um número um pouco maior de tentativas.

- Independência de Variáveis: Variáveis como tempo de resolução (time_to_complete_sec) e sequência de dias (streak_day) apresentam correlação próxima a 0,00 com todas as outras, provando que o tempo gasto ou os dias seguidos jogando não ditam se o usuário jogará melhor ou pior."
